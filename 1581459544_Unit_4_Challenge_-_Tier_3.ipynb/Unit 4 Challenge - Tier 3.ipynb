Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit.

```python
#I want to make sure that the data in each category is equal, so I'll use .count() again.
df_noNaN.count()
```

```python
#With the "Unnamed" values deleted, we also need to get rid of the extra columns. I will make a list of these that
#I will use to delete from the df_noNaN dataframe using the negation of .isin()
not_a_borough = ['Inner London', 'Outer London', 'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 'EAST MIDLANDS', 'WEST MIDLANDS', 'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 'SOUTH WEST', 'England']
df_noNaN = df_noNaN[~df_noNaN.London_Boroughs.isin(not_a_borough)]
df_noNaN.head()
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
#I'll subset the data on City of London and use a line plot to look at the month versus average price data.
london_aveprice = df[df['London_Boroughs'] == 'City of London']
price_chart = london_aveprice.plot(kind = 'line', x = 'Month', y = 'Average_price')
price_chart.set_ylabel('Average Price')
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column. 

To this end, you *could* apply a ***lambda function***. Your logic could work as follows:

```python
#To extract the year data from the dataframe, I'll use the lamba function and add a new column for year.
df['Year'] = df['Month'].apply(lambda t: t.year)
df.head()
```

```python
#I do want to make sure we analyze which boroughs have the greatest house price increase, on average, over the past two decades,
#so I should also group the boroughs and years together
df_grouped = df.groupby(by = ['London_Boroughs', 'Year']).mean()
df_grouped.head()
```

```python
#As I've done earlier, we need to make sure our index values are not our data, so I'll reset the index again.
df_grouped = df_grouped.reset_index()
df_grouped.head()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#This is my function to find the ratio of house prices between 2018 and 1998.
def averageprice_ratio(i):
    year_1998 = float(i['Average_price'][i['Year'] == 1998])
    year_2018 = float(i['Average_price'][i['Year'] == 2018])
    divide = [year_2018/year_1998]
    return divide

#To save this data, I'll store it in an empty dictionary with the key as the borough and the value as the ratio.
ratios = {}

#Loop through each borough and calculate the ratio
for j in df_grouped['London_Boroughs'].unique():
    borough = df_grouped[df_grouped['London_Boroughs'] == j]
    ratios[j] = averageprice_ratio(borough)

print(ratios)
```

```python
#I'll now convert the dictionary to a dataframe and check the results
df_final = pd.DataFrame(ratios)
df_final.head()
```

```python
#I want to transpose this data
df_final = df_final.T
df_final.head()
```

```python
#Again, I need to reset the index
df_final = df_final.reset_index()
df_final.head()
```

```python
#Since this is the ratio for 2018, I want to rename that column and the index column to reflect the boroughs listed.
df_final.rename(columns = {'index':'Borough', 0:'2018'}, inplace = True)
df_final.head()
```

```python
#Lets look at the ratios in descending order.
price_change = df_final.sort_values(by = '2018', ascending = False)
print(price_change)
```

```python
#I can also plot these values to better visualize the data.
price_ratio_plot = price_change[['Borough', '2018']].plot(kind = 'bar')
price_ratio_plot.set_xticklabels(price_change.Borough)
```
